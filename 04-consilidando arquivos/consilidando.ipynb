                        CONSOLIDANDO ARQUIVOS DE UMA PASTA    

```python
import os #importando a biblioteca os para manipulação de arquivos e diretórios
diretorio = (r"E:\Material de estudo\Arquivos\Origem\Consolidar")
dados = []
for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.xlsx'):
        caminho_arquivo = os.path.join(diretorio, arquivo)
        df = pd.read_excel(caminho_arquivo, engine='openpyxl')
        dados.append(df)
        
consolidado = pd.concat(dados, ignore_index=True)
consolidado.head(3)
```

                   SALVANDO CONSOLIDADO EM UMA  PASTA ESPECIFICA

```python
import os
pasta_destino = r'E:\Material de estudo\Arq-consolidados'#caminho da pasta onde o arquivo consolidado será salvo
nome_arquivo = "ConsolidadoVendas.xlsx" #nome do arquivo consolidado
caminho_completo = pasta_destino + '\\' + nome_arquivo#concatena o caminho da pasta com o nome do arquivo
consolidado.to_excel(caminho_completo, index=False, engine='openpyxl')# salva o DataFrame consolidado em um arquivo Excel
```

```python
pasta_destino = r'E:\Material de estudo\Arq-consolidados'#caminho da pasta onde o arquivo consolidado será salvo
nome_arquivo = "ConsolidadoVendas.csv" #nome do arquivo consolidado
caminho_completo = pasta_destino + '\\' + nome_arquivo#concatena o caminho da pasta com o nome do arquivo
consolidado.to_csv(caminho_completo, index=False,sep=',')# salva o DataFrame consolidado em um arquivo CSV
```
