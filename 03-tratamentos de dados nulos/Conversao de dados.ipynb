```python
import pandas as pd#importando pandas
```

```python
df = pd.read_csv(r'E:\Material de estudo\Arquivos\Origem\arquivos_csv\Clientes.csv',delimiter=',')#lendo o arquivo csv
df.head()
```

```python
                                           #Substituindo Valores nulos
df['street']=       df['street'].fillna('Não cadastrado')#preenchendo os valores nulos com string vazia
df['email']=        df['email'].fillna('Email cadastrado')#preenchendo os valores nulos com string vazia
df['number']=       df['number'].fillna('S/Numero')#preenchendo os valores nulos com string vazia
df['additionals']=  df['additionals'].fillna('S/Inf')#preenchendo os valores nulos com string vazia
df['state']=  df['state'].fillna('S/estado')#preenchendo os valores nulos com string vazia
df
```

```python
df['state']= df['state'].replace('São Paulo', 'SP ')#substituindo o nome do estado por sigla
df.head(5)
```

```python
# fazendo a substituição de valores
Valor_original='São Paulo'
Valor_substituto='SP'
df['state'] = df['state'].replace(Valor_original, Valor_substituto) #substituindo o nome do estado por sigla
df.head(5)
```

```python
mapeamento = {
    'São Paulo': 'SP',
    'Rio de Janeiro': 'RJ',
    'Minas Gerais': 'MG',
    'Rio Grande do Sul': 'RS',
    'Paraná': 'PR',
    'Santa Catarina': 'SC',
    'Bahia': 'BA',
    'Pernambuco': 'PE',
    'Ceará': 'CE',
    'Goiás': 'GO',
    'Acre' : 'AC',
    'Mato Grosso': 'MT',

} #dicionário de mapeamento para substituição de estados
df['state'] = df['state'].replace(mapeamento) #substituindo os nomes dos estados pelas siglas
df.head(5) #mostrando as primeiras 5 linhas do DataFrame
```
