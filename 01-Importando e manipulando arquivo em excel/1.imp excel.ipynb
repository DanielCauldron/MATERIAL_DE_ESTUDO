```python
# import pandas para manipulação de dados(pd significa pandas)
import pandas as pd
```

```python
# importando o arquivo excel df eo nome da variavel que vai receber o arquivo
df = pd.read_excel("E:/Material de estudo/Arquivos/Origem/arquivos_excel/Produto.xlsx", engine='openpyxl')# engine='openpyxl' é necessário para ler arquivos .xlsx
df.head(10)# Exibe as primeiras 5 linhas do DataFrame df
```

```python
#localizar a coluna 'Produto' e exibir os valores especificos
df_Produtos = df.loc[df['Name'] == 'Bolsa mais que Velha']# df_Produtos é variavel df.loc[df['Name']'= 'Bolsa mais que Velha'] filtra o DataFrame df para exibir apenas as linhas onde a coluna 'Name' é igual a 'Bolsa mais que Velha'
print(df_Produtos)# Exibe o DataFrame filtrado
# Exibe o DataFrame filtrado com a coluna 'Name' igual a 'Bolsa mais que Velha'
```

```python

df_Produtos = df.loc[df['Name'] == 'Bolsa mais que Velha', 'Price']# Filtra o DataFrame df para exibir apenas a coluna 'Price' onde a coluna 'Name' é igual a 'Bolsa mais que Velha'
print(df_Produtos)# Exibe o DataFrame filtrado com a coluna 'Price' onde a coluna 'Name' é igual a 'Bolsa mais que Velha'
# Exibe o DataFrame filtrado com a coluna 'Price' onde a coluna 'Name' é igual a 'Bolsa mais que Velha'
```

```python

itens = ['Calça','TV','Tênis', 'Perfume', 'Creme de Barbear','Shampoo']# Cria uma lista de itens
df_itens = df.loc[df['Name'].isin(itens)]# Filtra o DataFrame df para exibir apenas as linhas onde a coluna 'Name' está na lista de itens
df_itens#da Exibe o DataFrame filtrado com os itens da lista
```

```python
#localizar valores > ou < 1000 na coluna 'Price' e exibir os valores especificos
df_Maior1000 = df.loc[df['Price'] > 1000].sort_values(by='Price', ascending=False)# Filtra o DataFrame df para exibir apenas as linhas onde a coluna 'Price' é maior que 1000
df_Menor1000 = df.loc[df['Price'] < 1000].sort_values(by='Price', ascending=True )# Filtra o dataFrame df para exibir apenas as linhas onde a coluna 'Price'é menor que 1000
#Exibe as primeiras 5 linhas do Data Frame df_Mais1000
```
